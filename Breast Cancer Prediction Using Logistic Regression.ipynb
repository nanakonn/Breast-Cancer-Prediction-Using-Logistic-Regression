```python
# Data cleaning

#Removing Unnecesarry Column

df.drop(["id","Unnamed: 32"], axis=1, inplace=True)
```

Data is balanced.

After we found out that the data is balanced, now we moved on into :

1. Label the diagnosis column as a label data to separate that from the dataset 
2. As the result of many features on the dataset we cannot rely on all of that features, and now we have to decide which features is the best to implement on this case to reducec the data 
3. plotting heatmap to find coefficient or multicollinearity

```python
# as the result of labeling diagnosis column, we have to make diagnosis column into a intege

df["diagnosis"] = df["diagnosis"].replace({"M" : 1})
df["diagnosis"] = df["diagnosis"].replace({"B" : 0})

df
```

```python
# Heat map to found out the correllation 


fig, ax = plt.subplots(figsize=(25, 20))
df_cor = sns.heatmap(df.corr(), annot=True, cmap="Blues", ax=ax)

```

Scaling is different, then we moved into normalization 

```python
# Partitioning into X and Y 

X = df.drop("diagnosis", axis = 1)
Y = df["diagnosis"]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_Scale = scaler.fit_transform(X)
print(X_Scale)
```

# Dealing with Data Numerosity 

Find out what the best Features are using PCA

```python
from sklearn.decomposition import PCA

pca = PCA()
PrincipalComponent = pca.fit_transform(X_Scale)
PrincipalComponent
```

plotting PCA to see how many features that are best to implement to prediction 

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Principal Components')
plt.ylabel('Variance Covered')
plt.title('PCA')
plt.show()
```

From the diagram above we can see that at 0.9 and 1.0 are the best fit for this data because the covariens is
between 90% - 100%, therefore we selected 13 PCA to features in this dataset

```python
pca = PCA(n_components= 13)
principalComponent = pca.fit_transform(X_Scale)
principalDf = pd.DataFrame(data = principalComponent, 
                           columns = ["PC1","PC2","PC3","PC4","PC5","PC6","PC7","PC8","PC9","PC10","PC11","PC12", "PC13"])
principalDf
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test = train_test_split(principalDf,Y,test_size = 0.25, random_state= 50)

clf = LogisticRegression()
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred, digits = 2))
```
